```python
####### Creating Calculator app 

from tkinter import*
window=Tk()
```

```python
def addition():
    a=float(entry1.get())
    b=float(entry2.get())
    c=a+b
    entry3.insert(0,c)
```

```python
def subtraction():
    a=float(entry1.get())
    b=float(entry2.get())
    c=a-b
    entry3.insert(0,c)
```

```python
def multiplication():
    a=int(entry1.get())
    b=int(entry2.get())
    c=a*b
    entry3.insert(0,c)
```

```python
def division():
    a=int(entry1.get())
    b=int(entry2.get())
    c=a/b
    entry3.insert(0,c)
```

```python
 def modulo():
    a=int(entry1.get())
    b=int(entry2.get())
    c=a%b
    entry3.insert(0,c)   
```

```python
import math
```

```python
def tan_fun():
    a = float(entry1.get())
    b = math.tan(math.radians(a))
    entry3.insert(0, b)
```

```python
def sin_fun():
    a = float(entry1.get())
    b = math.sin(math.radians(a))
    entry3.insert(0,b)
```

```python
def cos_fun():
    a = float(entry1.get())
    b = math.cos(math.radians(a))
    entry3.insert(0,b)
```

```python
def delete():
    entry3.delete(0,END)
```
